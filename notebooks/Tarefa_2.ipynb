```python
imR = cv2.imread('estrela.jpg')
plt.imshow(imR)
plt.axis('off')
```

###### 2. Crie o seu próprio mapa de cores.
###### 3. Aplique o mapa na imagem colorida (em cada canal).
###### 4. Plote a imagem resultante da aplicação do mapa de cores.

```python
lut = np.zeros((256, 1, 3), dtype = np.uint8)
```

```python
for i in range(0, 256):
    lut[i, 0, 0] = (np.abs(i - 200)) % 256
    lut[i, 0, 1] = (np.abs(i - 130)) % 256
    lut[i, 0, 2] = (np.abs(i - 30)) % 256
```

```python
imgCM = cv2.applyColorMap(im, lut)
```

```python
plt.imshow(imgCM)
plt.axis('off')
```
