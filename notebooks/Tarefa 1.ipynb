```python
from cv2 import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
```

#### 1. Baixe uma imagem colorida qualquer (que seja poss√≠vel identificar uma pessoa ou um "objeto").
#### 2. Carregue a imagem para o formato RGB.

```python
img = cv2.imread('camera.jpg')
```

```python
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)
plt.axis('off')
```

#### 3. Separe os seus canais e exibe-os no script python.

```python
red = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
red[:, :, 1] = 0
red[:, :, 2] = 0
plt.imshow(red)
plt.axis('off')
```

```python
green = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
green[:, :, 0] = 0
green[:, :, 2] = 0
plt.imshow(green)
plt.axis('off')
```

```python
blue = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
blue[:, :, 0] = 0
blue[:, :, 1] = 0
plt.imshow(blue)
plt.axis('off')
```

#### 4. Junte apenas dois canais e renderize a imagem resultante. Para isso, um dos canais deve ter apenas valores iguais a zero.

```python
b, g, r = cv2.split(img)
channel_0 = np.zeros(img.shape[:2], dtype = 'uint8')
merged = cv2.merge([b,channel_0,r])
plt.imshow(merged)
plt.axis('off')
```

#### 5. Desenhe um c√≠rculo e ou um ret√¢ngulo para "marcar" uma pessoa ou um objetivo.
#### 6. Abaixo do que foi desenhado, escreva o que voc√™" marcou".

```python
imgR = img.copy()
cv2.rectangle(imgR, (80, 30), (260, 170), (255, 0, 0), 2)
cv2.putText(imgR, 'Claudi', (80, 70), cv2.FONT_HERSHEY_PLAIN, 3, (255, 0, 0), 2, cv2.LINE_AA)
plt.imshow(imgR)
plt.axis('off')
```
